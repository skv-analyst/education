- $x_i$ – аргумент; 
- $\epsilon_i$ – случайная величина с нулевым мат.ожиданием и некоторой постоянной дисперсией.

Значение $y_i$ прогнозируется с помощью следующей параметрической модели: $a(x,w) = w_0 + w_1 \cdot x$. 
- $w_0=0.5$;
- $w_1=-0.3$;
