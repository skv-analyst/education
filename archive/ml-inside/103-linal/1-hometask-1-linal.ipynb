__Вопрос: 7__  $f(x,y)=2x^2+y^4,∇f=?$
- $(a)\;\begin{pmatrix} 2x^2 \\ y^4 \end{pmatrix}$
- $(b)\;\begin{pmatrix} 4x \\ 4y^3 \end{pmatrix}$
- $(c)\; 4x + 4y^3$
- $(d)\;\begin{pmatrix} 4x \\ y^4 \end{pmatrix}$



__Ответ__. $(b)\;\begin{pmatrix} 4x \\ 4y^3 \end{pmatrix}$
<br>
1. Чтобы найти градиент функции, нужно вычислить частные производные функции по переменным $x$ и $y$
2. Найдем частную производную по $x$:
   $\frac{\partial f}{\partial x} = \frac{\partial}{\partial x} (2x^2 + y^4) = 4x$
3. Найдем частную производную по $y$:
   $\frac{\partial f}{\partial y} = \frac{\partial}{\partial y} (2x^2 + y^4) = 4y^3$   
4. Таким образом, градиент функции равен:
    $\nabla f = \begin{pmatrix} 4x \\ 4y^3 \end{pmatrix}$
