__Вопрос: 6__ $\frac{dln(x^2)}{dx}=?$
- $(a)\;x^{-2}$
- $(b)\;x^{-1}$
- $(c)\;2x^{-1}$
- $(d)\;2|x|^{-1}$


__Ответ:__ 
1. Перепишем задание в виде $\frac{d}{dx}ln(x^2)$
2. Используем свойство логарифма и вынесем степень за знак логарифма. $ln(x^2)=2ln(|x|)$
3. Теперь найдем производную и для этого используем правило дифференцирования логарифмической функции $ln(x) = \frac{1}{x}$, поэтому $$\frac{d}{dx}ln(x^2) = \frac{d}{dx}(2ln(|x|)) = 2 \cdot \frac{1}{|x|} \cdot \frac{d}{dx}(|x|) = \frac{2}{x} = 2x^{-1}$$ 
4. Итоговый ответ.  $(c)\; 2x^{-1}$

__Вопрос: 8__ Найдите $x$, при котором достигается минимум $f(x) = 3x^2 + 12x + 1$
- $(a)\;1$
- $(b)\;-1$
- $(c)\;2$
- $(d)\;-2$



__Ответ:__ 
1. Чтобы найти значение $x$ при котором достигается минимум функции, нужно найти точку в которой производная будет равна 0.
2. Найдем производную:$$f(x)'=\frac{d}{dx}(3x^2+12x+1)=\frac{d}{dx}(3x^2)+\frac{d}{dx}(12)+\frac{d}{dx}(1)=2\cdot3\cdot x^{2-1}+12+0=6x+12$$   
3. Решим уравнение в котором производная равна 0, $6x+12=0$ и получим $x=-2$
4. Итоговый ответ. $(d)\; -2$
