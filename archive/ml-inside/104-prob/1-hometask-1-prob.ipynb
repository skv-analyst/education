__Вопрос: 2__ Пусть модель для задачи классификации на классы 0 и 1 определила, что вероятность того, что заданный объект с вектором признаков $x$ принадлежит к классу $1$ равна $P(y = 1|x) = p_1$, а вероятность $P(y = 0|x) = p0 = 1−p_1$. Будем считать, что эти вероятности оценены моделью идеально точно. Модель в качестве прогноза класса $y$ для объекта $x$ выбирает $a(x) = arg\; \max_{s∈{0;1}}\; P(y = s|x)$ (говоря проще, тот класс, вероятность которого больше). Какова вероятность неверного ответа?
- $(a)\;max\left\{p_0, p_1 \right\}$
- $(b)\;min\left\{p_0, p_1 \right\}$
- $(c)\;p_0p_1$
- $(d)\;p_0 + p_1$


__Ответ.__ $(b)\;min\left\{p_0, p_1 \right\}$

Так как модель выбирает тот класс вероятность которого больше, значит вероятность правильного ответа равна max\left\{p_0, p_1 \right\}. Следовательно, вероятность неверного ответа будет $1-max\left\{p_0, p_1 \right\}$, а это равносильно записи $min\left\{p_0, p_1 \right\}$.

__Вопрос: 3__ Теперь в условиях предыдущей задачи рассмотрим модель, которая в качестве прогноза класса $y$ для объекта $x$ выбирает случайный ответ: $1$ с вероятностью $p1$ и $0$ с вероятностью $p0$. Какова будет вероятность неверного ответа теперь?
- $(a)\;2p_0p_1$
- $(b)\;min\left\{p_0, p_1 \right\}$
- $(c)\;p_0p_1$
- $(d)\;p_0+p_1$


__Ответ.__ $(a)\;2p_0p_1$
1. В данной задаче модель выбирает класс случайным образом: класс $1$ с вероятностью $p_1$, класс $0$ с вероятностью $1-p_1$
2. Рассчитаем вероятности возможных ошибок:
    - Вероятность того, что истинный класс $1$, а модель выбрала $0$: $p_1\cdot p_0$
    - Вероятность того, что истинный класс $0$, а модель выбрала $1$: $p_0\cdot p_1$
3. Суммируя полученные вероятности найдем итоговый ответ: $p_1\cdot p_0+p_0\cdot p_1=2p_0p_1$

__Вопрос: 5__ Симметричную монетку подбросили 5 раз, какова вероятность того, что при этом орел выпал 2 раза? 
- $(a)\;\frac{5}{12}$
- $(b)\;\frac{5}{16}$
- $(c)\;\frac{2}{5}$
- $(d)\;\frac{3}{32}$


__Ответ.__ $(b)\;\frac{5}{16}$
1. Для решения задачи используем формула Бернулли $P_{n}^{m}=C_{n}^{m}\cdot p^{m} \cdot q^{n-m}$
2. Подставим значения $P_{5}^{2}=C_{5}^{2}\cdot 0.5^{2} \cdot 0.5^{3}=0.3125=\frac{5}{16}$

__Вопрос: 6__ Симметричную монетку подбросили $7$ раз, $p_3$ - вероятность того, что при этом орел выпал $3$ раза, а $p_4$ - вероятность того, что при этом
орел выпал $4$ раза, тогда $\frac{p_3}{p_4}$?
- $(a)\;\frac{3}{4}$
- $(b)\;\frac{4}{3}$
- $(c)\;1$
- $(d)\;2$


__Ответ.__ $(c)\;1$.
Для решения этой задачи тоже используем формулу Бернулли, только теперь добавим чуть-чуть магии Python, чтобы не считать руками :)
--- 

```python
from math import comb

def bernoulli_formula(n, m, p):
    """
    Вычисляет вероятность того, что произойдет ровно m успехов в n испытаниях,
    при вероятности успеха p для каждого испытания.
    
    :param n: Общее количество испытаний
    :param m: Число успехов, которое нас интересует
    :param p: Вероятность успеха в одном испытании
    :return: Вероятность ровно m успехов в n испытаниях
    """
    # Вычисляем вероятность по формуле Бернулли
    probability = comb(n, m) * (p ** m) * ((1 - p) ** (n - m))
    return probability
```

```python
n = 7
p = 0.5
p3 = bernoulli_formula(n, 3, p)
p4 = bernoulli_formula(n, 4, p)

print(f'Вероятность p_3: {p3}\nВероятность p_4: {p4}\nВероятность p3/p4: {p3/p4}')
```

__Вопрос: 7__ Симметричную монетку подбросили $n$ раз ($n$ - четное), какова вероятность того, что при этом орел выпал $n/2$ раз?
- $(a)\;\frac{C_{n}^{n/2}}{2^n}$
- $(b)\;\frac{C_{n}^{\frac{n}{2}}}{2^{n/2}}$
- $(c)\;\frac{2^n}{C_{n}^{n/2}}$
- $(d)\;\frac{2^{n/2}}{C_{n}^{n/2}}$


__Ответ.__ $(a)\;\frac{C_{n}^{n/2}}{2^n}$
1. Опять берем формулу Бернулли и подставляем в нее значения: $P_{n}^{n/2}=C_{n}^{n/2}\cdot p^{n/2} \cdot q^{n-n/2}$
2. Так как у нас $p$ = $q$ = $1/2$, то можно $p^{n/2} \cdot q^{n-n/2}$ переписать как $\frac{1}{2}^{n/2} \cdot \frac{1}{2}^{n-n/2}=\frac{1}{2}^{n/2+(n-n/2)}=\frac{1}{2}^{n}$
3. Итого, получаем $P_{n}^{n/2}=C_{n}^{n/2}\cdot \frac{1}{2}^{n}=\frac{C_{n}^{n/2}}{2^n}$

__Вопрос: 8__ Половину пользователей отправили в группу А, в которой работал старый алгоритм показа рекламы, а половину пользователей отправили в группу В, в которой работал новый алгоритм показа рекламы. Спустя 7 дней выяснилось, что алгоритм B проиграл по конверсии показа баннера в клик на баннер всего в 1 день из 7. Если представить, что между алгоритмами не было разницы по эффективности и каждый день побеждал либо алгоритм А с вероятностью 50%, либо алгоритм В с вероятностью 50%, то с какой вероятностью могло оказаться, что В проиграл только в 1 день из 7? Округляйте результат до десятых долей процента.
- $(a)\;2.5%$
- $(b)\;4.5%$
- $(c)\;5.5%$
- $(d)\;7.5%$


__Ответ.__ $(c)\;5.5%$

2. Подставим значения и посчитаем $C_{7}^{1}\cdot 0.5^{1} \cdot 0.5^{7-1}=0.054=5.46%$
