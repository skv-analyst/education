```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(seed=42)
population = rng.normal(size=1000, loc=25, scale=3)
sns.displot(population, height=4);
```

```python
def diff_mean_ab(data):
    a = np.mean(np.random.choice(data, size=2))
    b = np.mean(np.random.choice(data, size=2))
    return a - b
```

```python
res = []
for i in range(100000 + 1):
    res.append(diff_mean_ab(data=population))
sns.displot(res);
```
